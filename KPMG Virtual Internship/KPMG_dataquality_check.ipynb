```python
#deceased_indicato refers to the people who are alive and not alive. 
df3.deceased_indicator.value_counts()
```

```python
#We need the data of only the people who are alive so lets remove the row values with 'Y'
df3 = df3[df3["deceased_indicator"].str.contains("Y") == False]
```
